### PROGRAMMING QUESTION
Question: How can you handle duplicate values in a dataset for a variable in Python using pandas?
Suppose, you are given the following dataset:
df = pd.read_csv('file.csv')
This dataset has many duplicate values. You need to identify them and also remove them.

### Importing Needed packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

### Loading the data into the DataFrame

```python
df = pd.read_csv('file.csv')
```

```python
# To display the top 5 rows

df.head()
```

```python
# To display the bottom 5 rows

df.tail()
```

### Dropping the duplicate rows

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]

print("Number of duplicate rows: " , duplicate_rows_df.shape)
```

### Now  let us remove the duplicate data

```python
df.count() # Used to count the number of rows
```

```python
df = df.drop_duplicates()
```

```python
df.head()
```

```python
df.count()
```

### Question: Explain dimension reduction in machine learning.

Answer: Dimension Reduction refers to the process of converting a set of data having vast dimensions into data with lesser dimensions ensuring that it conveys similar information concisely. These techniques are typically used while solving machine learning problems to obtain better features for a classification or regression task.
